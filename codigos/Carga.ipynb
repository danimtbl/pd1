```python
#imports necess√°rios
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession

```

```python
format = 'json'
inferSchema = 'false'
header = 'false'
multiline = 'false'
```

```python
#Configurando spark
conf = SparkConf().setAppName('Load').set("spark.serializer", "org.apache.spark.serializer.KryoSerializer").set("hive.exec.dynamic.partition", "true").set("hive.exec.dynamic.partition.mode", "nonstrict")
spark = SparkSession.builder.config(conf=conf).enableHiveSupport().getOrCreate()
```

```python
#Definindo caminho dos arquivos
filePath1 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\students.json'
filePath2 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\courses.json'
filePath3 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\subjects.json'
filePath4 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\universities.json'
filePath5 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\subscriptions.json'
filePath6 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\sessions.json'
filePath7 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE A\\student_follow_subject.json'
filePath8 = 'C:\\Users\\dbello\Documents\\pd-desafio\\Datasets - Teste Data Engineer - Passei Direto\\BASE B\\*.json'
```

```python
#lendo arquivo para dataframe para descobrir schema de cada json
df_students = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath1)
df_courses = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath2)
df_subjects = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath3)
df_universities = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath4)
df_subscriptions = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath5)
df_sessions = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath6)
df_student_follow_subject = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath7)
df_baseB = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath8)
```

```python
#descobrindo o schemas de cada json
df_students.printSchema()
df_courses.printSchema()
df_subjects.printSchema()
df_universities.printSchema()
df_subscriptions.printSchema()
df_sessions.printSchema()
df_student_follow_subject.printSchema()
df_baseB.printSchema()
```

```python
#Definindo schema de cada df
schema_students = StructType([
    StructField("City", StringType(), True),
    StructField("CourseId", LongType(),True),
    StructField("Id", StringType(), True),
    StructField("RegisteredDate", StringType(), True),
    StructField("SignupSource", StringType(), True),
    StructField("State", StringType(), True),
    StructField("StudentClient", StringType(), True),
    StructField("UniversityId", LongType(),True)])

schema_courses = StructType([
    StructField("Id", LongType(),True),
    StructField("Name", StringType(), True)])

schema_subjects = StructType([
    StructField("Id", LongType(),True),
    StructField("Name", StringType(), True)])

schema_universities = StructType([
    StructField("Id", LongType(),True),
    StructField("Name", StringType(), True)])

schema_subscriptions = StructType([
    StructField("PaymentDate",StringType(), True),
    StructField("PlanType",StringType(), True),
    StructField("StudentId", StringType(), True)])

schema_sessions = StructType([
    StructField("SessionStartTime",StringType(), True),
    StructField("StudentClient",StringType(), True),
    StructField("StudentId", StringType(), True)])

schema_student_follow_subject = StructType([
    StructField("FollowDate",StringType(), True),
    StructField("StudentId",StringType(), True),
    StructField("SubjectId", LongType(), True)])

schema_baseB = StructType([
    StructField("Last Accessed Url", StringType(), True),
    StructField("Page Category", StringType(),True),
    StructField("Page Category 1", StringType(), True),
    StructField("Page Category 2", StringType(), True),
    StructField("Page Category 3", StringType(), True),
    StructField("Page Name", StringType(), True),
    StructField("at", StringType(), True),
    StructField("browser", StringType(), True),
    StructField("carrier", StringType(), True),
    StructField("city_name", StringType(), True),
    StructField("clv_total", LongType(), True),
    StructField("country", StringType(), True),
    StructField("custom_1", StringType(), True),
    StructField("custom_2", StringType(), True),
    StructField("custom_3", StringType(), True),
    StructField("custom_4", StringType(), True),
    StructField("device_new", BooleanType(), True),
    StructField("first-accessed-page", StringType(), True),
    StructField("install_uuid", StringType(), True),
    StructField("language", StringType(), True),
    StructField("library_ver", StringType(), True),
    StructField("marketing_campaign", StringType(), True),
    StructField("marketing_medium", StringType(), True),
    StructField("marketing_source", StringType(), True),
    StructField("model", StringType(), True),
    StructField("name", StringType(), True),
    StructField("nth", LongType(), True),
    StructField("os_ver", StringType(), True),
    StructField("platform", StringType(), True),
    StructField("region", StringType(), True),
    StructField("session_uuid", StringType(), True),
    StructField("studentId_clientType", StringType(), True),
    StructField("type", StringType(), True),
    StructField("user_type", StringType(), True),
    StructField("uuid", StringType(),True)])
```

```python
#lendo arquivo para dataframe com schemas definidos
df_students = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath1, schema_students).repartition(1)
df_courses = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath2, schema_courses).repartition(1)
df_subjects = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath3, schema_subjects).repartition(1)
df_universities = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath4, schema_universities).repartition(1)
df_subscriptions = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath5, schema_subscriptions).repartition(1)
df_sessions = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath6, schema_sessions).repartition(1)
df_student_follow_subject = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath7, schema_student_follow_subject).repartition(1)
df_baseB = spark.read.format(format) \
                    .option("inferSchema", inferSchema) \
                    .option("header", header) \
                    .option("multiLine", multiline) \
                    .json(filePath8, schema_baseB).repartition(3)
```
